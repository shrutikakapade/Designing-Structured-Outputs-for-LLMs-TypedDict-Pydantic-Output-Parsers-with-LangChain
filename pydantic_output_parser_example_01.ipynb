```python
from langchain_core.output_parsers import PydanticOutputParser
```

```python
from pydantic import BaseModel
```

```python
# Create Data format

class student_project_review(BaseModel):
    name:str
    batch:int
    Branch:str
    project_review_status:str
    total_marks:int
```

```python
#create object for output parser

parser = PydanticOutputParser(pydantic_object=student_project_review)
```

```python
# create prompt template

from langchain_core.prompts import ChatPromptTemplate
```

```python
final_prompt = ChatPromptTemplate.from_messages([
    ("system", "You have a 5 + experience in data extraction"),
    ("human",'''Extract the student information from the given text
    text: {input data}
    {format_instructions}''')
])
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
import os
```

```python
from google.colab import userdata
gen = userdata.get('Gemini')
```

```python
os.environ["GOOGLE_API_KEY"]=gen
```

```python
llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")
```

```python
# Use chain concept pipeline of[prompt,LLM,Output parser] connect them together
chain = final_prompt | llm | parser
```

```python
chain.invoke({
    "input data": '''student name is shrutika kapade, she is from 361 batch,
    from pune branch. project review status is Yes and she got total marks 40/100.
    feedback: she need to improve technical skills and communication skills.
    ''',
    "format_instructions":parser.get_format_instructions()
})
```
